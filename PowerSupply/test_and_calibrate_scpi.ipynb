```python
#VOLT ADC CALIBRATION
device.write('OUTP ON')
sleep(0.2)
device.write('FUNC VOLT')
sleep(0.2)
device.write('SOUR:VOLT 0')
sleep(5)
calibration_table = []
TARGET = 1.5
v_set = 0

def safe_query(inst, cmd):
    try:
        val = inst.query(cmd).strip()
        return float(val)
    except Exception as e:
        print("SCPI error:", e)
        return None

while True:
    v_now = safe_query(dmm, 'MEAS:VOLT?')
    
    if v_now is None:
        sleep(0.2)
        continue

    print(f"\rVoltage DMM [V]: {v_now:.5f}", end="")

    if v_now >= TARGET:
        print("\nReached target voltage")
        break

    # --- take 3 averaged samples ---
    dmm_vals = []
    adc_vals = []
    for _ in range(5):
        vcc = safe_query(dmm, 'MEAS:VOLT?')
        v_adc = safe_query(device, 'MEAS:VOLT?')

        if v_adc is not None and vcc is not None:
            dmm_vals.append(vcc)
            adc_vals.append(v_adc)

        sleep(0.3)  # keep this (manual adjustment time)


    vcc_avg = round(sum(dmm_vals) / len(dmm_vals),7)
    adc_avg = round(sum(adc_vals) / len(adc_vals),7)

    calibration_table.append((round(adc_avg,7), round(vcc_avg,7)))

    print(f"\nSaved point: ADC={adc_avg:.7f}, DMM={vcc_avg:.7f}")

    sleep(1.5)  # time for you to adjust knob
    if v_set < 0.5:
        v_set += 0.005
    else:    
        v_set += 0.01
    device.write('SOUR:VOLT '+str(v_set))

print("\nCalibration table:")
print(calibration_table)
```

```python
#CURRENT DAC CALIBRATION
device.write('OUTP ON')
sleep(0.2)
device.write('FUNC CURR')
sleep(0.2)
device.write('SOUR:CURR 0')
sleep(5)
calibration_table = []
TARGET = 150
i_set = 0

def safe_query(inst, cmd):
    try:
        val = inst.query(cmd).strip()
        return float(val)
    except Exception as e:
        print("SCPI error:", e)
        return None


while True:
    i_now = safe_query(dmm, 'MEAS:CURR?')*1000

    if i_now is None:
        sleep(0.2)
        continue

    print(f"\rCurrent mA: {i_now:.3f}", end="")

    if i_now >= TARGET:
        print("\nReached target current")
        break

    # --- take 3 averaged samples ---
    dmm_vals = []

    for _ in range(5):
        icc = safe_query(dmm, 'MEAS:CURR?')*1000

        if icc is not None:
            dmm_vals.append(icc)

        sleep(0.1)  # keep this (manual adjustment time)

    icc_avg = sum(dmm_vals) / len(dmm_vals)

    calibration_table.append((round(icc_avg,7), round(i_set,7)))

    print(f"\nSaved point: DMM={icc_avg:.4f}, DAC={i_set:.4f}")

    sleep(1.5)  # time for you to adjust knob
    if i_set < 5:
        i_set += 0.1
    elif i_set < 25:
        i_set += 0.5
    else:    
        i_set += 1

    device.write('SOUR:CURR '+str(i_set))

print("\nCalibration table:")
print(calibration_table)
```

```python
#LIMIT VOLT CALIBRATION
device.write('FUNC VOLT')
device.write('SOUR:VOLT 0')
sleep(5)
target_max = 5.60
calibration_table = []

def safe_query(inst, cmd):
    try:
        return float(inst.query(cmd).strip())
    except:
        return None

for limit in [round(x * 0.5, 2) for x in range(1, int(target_max*2)+1)]:

    print(f"\n--- Calibrating V limit: {limit}V ---")
    sleep(0.5)
    device.write(f"SOUR:VOLT:PROT {limit}")
    sleep(0.5)
    device.write("SOUR:VOLT 0")

    sleep(0.5)

    prev_v = 0
    stable_count = 0

    v_set = 0

    while True:
        v_set += 0.05
        device.write(f"SOUR:VOLT {v_set}")

        sleep(0.2)

        v_now = safe_query(dmm, "MEAS:VOLT?")
        if v_now is None:
            continue

        print(f"\rSet={v_set:.2f}  Meas={v_now:.3f}", end="")

        # Detect saturation (no longer increasing)
        if abs(v_now - prev_v) < 0.002:
            stable_count += 1
        else:
            stable_count = 0

        prev_v = v_now

        if stable_count >= 5:
            print(f"\nClamp detected at {v_now:.4f} V")
            calibration_table.append((v_now, limit))
            break

        if v_set > target_max + 1:
            print("\nSafety break")
            break

print("\nVoltage limit table:")
print(calibration_table)

```
